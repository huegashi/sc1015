```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt

housedata = pd.read_csv('train.csv')
housedata.head()
```

```python
housedata.dtypes
```

```python
intdf=housedata.select_dtypes(include = 'int64')
intdf
```

```python
intdf.drop(columns=['MSSubClass', 'OverallQual', 'OverallCond'])
```

```python
numdata = pd.DataFrame(intdf)
numdata.describe().round(2)
```

```python
spsum = pd.DataFrame(intdf['SalePrice'])
spsum.describe().round(2)
```

```python
f = plt.figure(figsize=(16,8))
sb.boxplot(data = numdata['SalePrice'], orient = "h")
```

```python
f = plt.figure(figsize=(16,8))
sb.histplot(data = numdata['SalePrice'])
```

```python
f = plt.figure(figsize=(16,8))
sb.violinplot(data = numdata['SalePrice'], orient = "h")
```
